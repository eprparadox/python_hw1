```python
## take the image and the enlargement factor
def enlarge_image(image, factor):
    ## intialize an array of zeros to fill with the info
    new_image = zeros((image.shape[0]*factor,image.shape[1]*factor))

    ## ugh, i know i shouldn't be using this counter but ...
    counter = 0
    ## now grab each row ...
    for row in range(image.shape[0]):
        old_row = image[row,:]
        new_row = []
        
        ## and copy each value in the old row
        ## factor times
        for val in old_row:
            for _ in itertools.repeat(None,factor):
                new_row.append(val)
        for _ in itertools.repeat(None,factor):
            new_image[counter,:] = new_row
            counter += 1
    return new_image                    

        
```

**Solution to 1-0** The first answer wants all the offsets.  Load up each image, enlarge, and calculate offset using the cross correlation (as per the hint).  I found the formula to actually calculate offset in this (ironic) matlab page : http://www.mathworks.com/help/images/examples/registering-an-image-using-normalized-cross-correlation.html

```python
## first, make a cropping function
def crop_image(image,val):
    ## crop an image on rows where 
    ## the entire row is equal to
    ## the val val
    to_remove = []
    
    for row in range(image.shape[0]):
        ## if the whole row is 255 
        test = image[row,:] - np.ones(len(image[row,:])) * val
        if not test.any():
            to_remove.append(row)
        
    cropped_image = numpy.delete(image, to_remove, axis=0)    
    return cropped_image
```

```python
import scipy 
from scipy import signal

## intialize list of offsets
offsets = []

## first image is the first one we're working with
template = imread(image_list[0],flatten=True)
template = enlarge_image(template,3)
med_val = median(template)

## crop and thresh template
template[template < med_val] = 0; 
template = crop_image(template,255)

## maybe ... find a center first?
cross_corr = scipy.signal.fftconvolve(template,template[::-1, ::-1],mode='same')
## get the max val and find the subscripts
xcen, ycen = np.unravel_index(cross_corr.argmax(), template.shape)

## load up each image, get the offset and add to list
for idx,image in enumerate(image_list):
    im = imread(image,flatten=True)
    ## enlarge by factor of 3
    enlarged_image = enlarge_image(im,3)
    ## thresh and remove white space as per above
    enlarged_image[enlarged_image < med_val] = 0
    enlarged_image = crop_image(enlarged_image,255)
    ## get offset via cross correlation
    cross_corr = scipy.signal.fftconvolve(template,enlarged_image[::-1, ::-1],mode='same')
    ## get the max val and find the subscripts
    xcoord, ycoord = np.unravel_index(cross_corr.argmax(), template.shape)
    ## record
    offsets.append((xcoord-xcen,ycoord-ycen))
    
    
```

** Solution to 1-1**
If it is, we should be able to fit the rest of the assignment into the following code block.  We have the list of offsets so we load up each image, shift it, and pack them all into a 3d array.  Then we take the mean on the 3rd dimension and blammo: super resolution imaging.

```python
import numpy as np

## initilize array
the_dims = (template.shape[0],template.shape[1],len(image_list))
image_stack = np.zeros(the_dims)

for idx, (image, offset) in enumerate(zip(image_list, offsets)):
    
    ## load it up
    im = imread(image,flatten=True)
    ## enlarge by factor of 3
    enlarged_image = enlarge_image(im,3)
    ## crop by same criterion
    enlarged_image = crop_image(enlarged_image,255)
    
    ## shift
    shifted_image = scipy.ndimage.interpolation.shift(enlarged_image,offset,order = 3,mode='wrap')
    
    ## pack in zero array
    image_stack[:,:,idx] = shifted_image
    
```

```python
the_mean = np.mean(image_stack,axis=2)
```
