ok, that basically looks like a car.  grey is probably not the right color map, but we can worry about that later.  part one is to enlarge the image 3x on each axis.  i think we'll do this by writing an enlarging function that takes a factor and enlarges the image by that factor row by row.  for each row, we'll just for value by value across the row and copy the values that many times.

there's gotta be a more clever way to do this but this is pretty short.


```python
## take the image and the enlargement factor
def enlarge_image(image, factor):
    ## intialize an array of zeros to fill with the info
    new_image = zeros((image.shape[0]*factor,image.shape[1]*factor))

    ## ugh, i know i shouldn't be using this counter but ...
    counter = 0
    ## now grab each row ...
    for row in range(image.shape[0]):
        old_row = image[row,:]
        new_row = []
        
        ## and copy each value in the old row
        ## factor times
        for val in old_row:
            for _ in itertools.repeat(None,factor):
                new_row.append(val)
        for _ in itertools.repeat(None,factor):
            new_image[counter,:] = new_row
            counter += 1
    return new_image                    

        
```
